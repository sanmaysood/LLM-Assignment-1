```python
import torch
import warnings
from transformers import AutoModelForCausalLM, AutoTokenizer, GenerationConfig

warnings.filterwarnings('ignore')
```

```python
OH_model_name = 'sarvamai/OpenHathi-7B-Hi-v0.1-Base'

OH_model = AutoModelForCausalLM.from_pretrained(
    OH_model_name,
    load_in_4bit=True,
    device_map='auto'
)

OH_tokenizer = AutoTokenizer.from_pretrained(OH_model_name)
OH_tokenizer.pad_token = OH_tokenizer.eos_token

OH_model.config
```

```python
llama_model_name = 'meta-llama/Meta-Llama-3-8B-Instruct'

llama_model = AutoModelForCausalLM.from_pretrained(
    llama_model_name,
    load_in_4bit=True,
    device_map='auto'
)

llama_tokenizer = AutoTokenizer.from_pretrained(llama_model_name)
llama_tokenizer.pad_token = llama_tokenizer.eos_token

llama_model.config
```

```python
def generate_response(model, tokenizer, prompt):

    inputs = tokenizer(prompt, return_tensors="pt")
    input_ids = inputs["input_ids"].to('cuda')

    generation_config = GenerationConfig(
        do_sample = True,
        temperature = 0.4,
        repetition_penalty = 1.5,
        max_new_tokens = 150
    )

    with torch.no_grad():
        generation_output = model.generate(
            input_ids=input_ids,
            attention_mask=torch.ones_like(input_ids),
            generation_config=generation_config
        )

    response = tokenizer.decode(generation_output[0].cuda(), skip_special_tokens=True).strip()

    return response
```

```python
self_consistency_questions = [
    "You looked at the clock at 3:15. How many degrees are in the angle between the hour and minute hand?",
    "Michael is the father of Sarah. Sarah’s grandfather is George, who has two children, Michael and Lisa. Lisa is married to Paul, and they have a son named Mark. How is Mark related to Sarah?",
    "A rooster lays an egg on the roof of a barn. The roof is sloped equally on both sides. Which way will the egg roll?"
]

fact_checking_questions = [
    "Which Indian won two bronze medals at the 2024 Paris Olympics?",
    "Who won the Women's Singles US Open 2024?",
    "Which company acquired Splunk?"
]
```

```python
print("\nOpenHathi Self Consistency\n")
for question in self_consistency_questions:
    response1 = generate_response(OH_model, OH_tokenizer, question)
    response2 = generate_response(OH_model, OH_tokenizer, question)
    print(f"Question: {question}\n")
    print(f"Response 1: {response1}\n")
    print(f"Response 2: {response2}\n")
    print("-" * 200)


print("\nOpenHathi Fact Checking\n")
for question in fact_checking_questions:
    response = generate_response(OH_model, OH_tokenizer, question)
    print(f"Question: {question}\n")
    print(f"Response: {response}\n")
    print("-" * 200)
```

```python
import os
import random
import string
import pinecone
import warnings
import requests
from langchain import HuggingFacePipeline
from langchain.chains import RetrievalQA
from langchain.vectorstores import Pinecone
from langchain.embeddings.huggingface import HuggingFaceEmbeddings
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
from pinecone import Pinecone, PodSpec

warnings.filterwarnings('ignore')
```

```python
embedding_model_id = 'sentence-transformers/all-MiniLM-L6-v2'

embedding_model = HuggingFaceEmbeddings(
    model_name=embedding_model_id,
    model_kwargs={'device':'cuda'},
    encode_kwargs={'device': 'cuda', 'batch_size': 16}
)
```

```python
google_doc_url = "https://docs.google.com/document/d/1a3-JnlLPSBOwvmN0V9GosPmVfyUZ3Jj73kdsK_ZJahA/edit"
documents = []

export_url = google_doc_url.replace("/edit?usp=sharing", "/export?format=txt")
response = requests.get(export_url)
documents.append(response.text)
```

```python
pc = Pinecone(api_key='3d587eab-16df-40b5-9f5c-7b73c7ebae30')

index_name = 'llm-course-ass1-rag'

existing_indexes = pc.list_indexes().names()

if index_name not in existing_indexes:
    try:
        pc.create_index(
            index_name,
            dimension=384,
            metric='cosine',
            spec=PodSpec("gcp-starter")
        )
    except Exception as e:
        print(f"Error creating index: {e}")
else:
    print(f"Index '{index_name}' already exists.")

index = pc.Index(index_name)
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.docstore.document import Document
from langchain.vectorstores import Pinecone

text_splitter = RecursiveCharacterTextSplitter(chunk_size = 400, chunk_overlap = 40)
docs = [Document(page_content = content) for content in documents]
chunked_docs = text_splitter.split_documents(docs)


texts = [doc.page_content for doc in chunked_docs]
embeddings = embedding_model.embed_documents(texts)


for i, doc in enumerate(chunked_docs):
    doc_id = f"doc-{i}"
    embedding = embeddings[i]
    if embedding is not None and len(embedding) > 0:
      index.upsert(vectors = [(doc_id, embedding, {'text': doc.page_content})])
```

```python
pipe_llama = pipeline(
    'text-generation',
    model = llama_model,
    tokenizer = llama_tokenizer,
    device_map='auto',
    max_new_tokens=512,
    temperature=0.8,
    do_sample=True,
    top_k=10,
    num_return_sequences=1,
    eos_token_id= llama_tokenizer.eos_token_id
)
llm_llama = HuggingFacePipeline(pipeline=pipe_llama)

vectorstore = Pinecone(index, embedding_model.embed_query, "text")
```

```python
def print_answer(response):
  response_text = response['result']

  start_index = response_text.find("Final Answer:")
  if start_index != -1:
    final_answer = response_text[start_index + len("Final Answer:"):].split('\n')[0].strip()

  start_ind = response_text.find("Helpful Answer:")
  if start_ind != -1:
    helpful_answer = response_text[start_ind + len("Helpful Answer:"):].split('\n')[0].strip()

  if(start_index != -1):
    print("Final Answer:", final_answer)

  elif (start_ind != -1):
    print("Helpful Answer:", helpful_answer)

  print()
```

```python
rag_pipeline = RetrievalQA.from_chain_type(
    llm = llm_llama,
    chain_type='stuff',
    retriever=vectorstore.as_retriever()
)

print("\nLLAMA3 Self Consistency\n")
for query in self_consistency_questions:
  response_with_rag1 = rag_pipeline(query)
  response_with_rag2 = rag_pipeline(query)
  print(f"Question: {query}")
  print("Response 1\n")
  print_answer(response_with_rag1)
  print("Response 2\n")
  print_answer(response_with_rag2)
  print("-" * 200)

print("\nLLAMA3 Fact Checking\n")
for query in fact_checking_questions:
  response_with_rag = rag_pipeline(query)
  print(f"Question: {query}")
  print_answer(response_with_rag)
  print("-" * 200)
```
